```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm
from sklearn.model_selection import train_test_split
```

```python
DATA_PATH = 'data/train.csv'
TEST_PATH = 'data/test.csv'
ATTRS_PATH = 'data/attr.csv'
FIGSIZE = (16, 7)
RANDOM_STATE = 42
```

```python
sns.set()
tqdm.pandas()

pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
df = pd.read_csv(DATA_PATH,
    nrows=1500
)
attrs = pd.read_csv(ATTRS_PATH)
```

```python
joined_u = pd.merge(df, attrs, left_on=['ego_id', 'u'], right_on=['ego_id', 'u'], how='left')
joined_u
```

```python
joined_v = pd.merge(joined_u, attrs, left_on=['ego_id', 'v'], right_on=['ego_id', 'u'], how='left', suffixes=('_u', '_v'))
joined_v
```

```python
joined_v = joined_v[[
    'ego_id', 'u_u', 'v', 't', 'x1', 'x2', 'x3', 'age_u', 'city_id_u',
    'sex_u', 'school_u', 'university_u', 'age_v', 'city_id_v',
    'sex_v', 'school_v', 'university_v'
]]
joined_v.columns = [
    'ego_id', 'u', 'v', 't', 'x1', 'x2', 'x3', 'age_u', 'city_id_u',
    'sex_u', 'school_u', 'university_u', 'age_v', 'city_id_v',
    'sex_v', 'school_v', 'university_v'
]
```

```python
graph_ids = df['ego_id'].unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(graph_ids, range(len(graph_ids)), train_size=0.7, random_state=RANDOM_STATE)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, train_size=0.8, random_state=RANDOM_STATE)
```
